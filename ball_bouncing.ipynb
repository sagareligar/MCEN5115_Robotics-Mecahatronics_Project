```
# inspired by https://www.youtube.com/watch?v=N524t6wdlcM
# https://www.mathworks.com/help/symbolic/examples/validate-simulink-model-using-symbolic-math-toolbox.html

import numpy as np
import matplotlib.pyplot as plt

c_bounce = 1;        # Bouncing's coefficient of restitution; perfect restitution would be 1
gravity  = 9.8;       # Gravity's acceleration (m/s)
height_0 = 10;        # Initial height at time t=0 (m)
velocity_0=0;        # Initial velocity at time t=0 (m/s)


simTime=25 # simulation time (s)
tStep=0.001 # simulation time step
iterations=int(simTime/tStep) # total number of iterations
t=np.arange(0,iterations)


h=np.zeros((iterations,1))
h[1,:]=height_0
v=np.zeros((iterations,1))
v[1,:]=velocity_0

for i in range(1,iterations-1):
  v[i+1,:]=v[i,:] - (gravity*tStep)


  h[i+1,:]=h[i,:] + v[i,:]*tStep

  if(h[i+1,:]<0):
     v[i,:]= - (v[i,:] * c_bounce)
     v[i+1,:]=v[i,:] - (gravity*tStep)
     h[i+1,:]=h[i,:] + v[i,:]*tStep


plt.rcParams["figure.figsize"] = (15,10) # resizes figures for viewing
plt.figure()
plt.plot(t,h,'r',label='ball')

plt.ylabel('Height (m)')
plt.title('Bouncing_Ball')

  

```
